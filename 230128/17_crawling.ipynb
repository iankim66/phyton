```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(html_data, 'html.parser')
soup
```

-태크의 이름을 사용하여 검색 
  - tag : 해당 태그의 첫번째 정보를 출력 
  - tag[속성] : 해당 태그의 첫번째 속성에 대한 값을 출력

```python
print ( soup.title)
print ( soup.title.name)
print ( soup.title.string)
```

```python
print(soup.img)
```

```python
print(soup.img['alt'])
```

- find()
   - HTML 안에서 해당 Tag 에 대한 첫번째 정보를 출력 
   - find (속성 = '값') : HTML 해당 속성과 일치하다는 값에 대한 첫번째 정보를 출력 
- find_all () 
   - HTML 안에 해당 TAG 에 대한 모든 정보를 list 형식으로 출력.
   - Limit 속성 : 리스트의 길이 지정 

```python
print(soup.find('a'))  ## 'a' 라는 속성을 가진  첫번째 하이퍼 링크를 찾아서 보여줌
```

```python
print(soup.find(id = 'newsstand'))  ## id = newsstand 속성을 가진것을 전부 보여줌 
```

```python
print(soup.find_all('span')) ## span 으로 시작된느 값들을 list 형태로 가져와서 보여줌.
```

```python
print(soup.find_all('span')[2])  ## 2번째의 span 값을 가져옴
```

```python
sample = soup.find('div', attrs={
    'class' : 'list_theme_wrap'
 })
sample
```

```python
sample_list = sample.find_all('strong')
```

```python
for i in sample_list:
    print(i.get_text())
```
